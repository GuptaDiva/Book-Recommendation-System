```python
import pandas as pd

def explode_genres(df):
    # Split genres into a list for each row
    df['genres'] = df['genres'].apply(lambda x: eval(x))  # Convert the string representation to a set or list
    # Explode the DataFrame so each genre gets its own row
    df = df.explode('genres').reset_index(drop=True)
    return df

# Apply the function to your DataFrame
df_expanded = explode_genres(final_books)

# Display the new DataFrame with separated genres
print(df_expanded)
```
